```python
df_cm = pd.read_csv('commits_data.csv')
```

```python
df_cm['committer_username'].unique()
df_cm = df_cm[(df_cm['committer_username'] != 'gungui98') & (df_cm['committer_username'] != 'lqanh11')]
```

```python
df_cm.sort_values('commit_date', ascending = False)
```

```python
df_cm['commit_date'] = pd.to_datetime(df_cm['commit_date'])
df_cm['commit_date_by_date'] = df_cm['commit_date'].dt.date
df_cm['commit_date_by_time'] = df_cm['commit_date'].dt.hour
```

```python
df_cm['commit_date']
```

```python
df_commit_by_date = df_cm[['commit_sha', 'commit_date_by_date']].sort_values('commit_date_by_date').groupby('commit_date_by_date').count().reset_index()
```

```python
df_commit_by_date
```

```python
fig_cm = df_commit_by_date.plot('commit_date_by_date', 'commit_sha', figsize = (15, 5))
fig_cm.axvline("2023-04-28", color="red", linestyle = "dashed")
fig_cm.axvline("2023-05-05", color="red", linestyle = "dashed")
fig_cm.axvline("2023-05-08", color="red", linestyle = "dashed")
fig_cm.axvline("2023-05-11", color="red", linestyle = "dashed")
fig_cm.axvline("2023-05-12", color="red", linestyle = "dashed")
fig_cm.axvline("2023-05-14", color="red", linestyle = "dashed")
fig_cm.axvline("2023-05-19", color="red", linestyle = "dashed")
fig_cm.axvline("2023-05-22", color="red", linestyle = "dashed")
```

```python
df_commit_by_time = df_cm[['commit_sha', 'commit_date_by_time']].sort_values('commit_date_by_time').groupby('commit_date_by_time').count().reset_index()
df_commit_by_time['commit_date_by_time'] = df_commit_by_time['commit_date_by_time'].astype(int)
df_commit_by_time
```

```python
for i in range(24):
  if i not in df_commit_by_time['commit_date_by_time'].values:
    add_row = pd.DataFrame([{
        'commit_date_by_time' : i, 
        'commit_sha' : 0
        }])
    df_commit_by_time = pd.concat([df_commit_by_time, add_row])
df_commit_by_time = df_commit_by_time.sort_values('commit_date_by_time')
```

```python
df_commit_by_time
```

```python
fig = df_commit_by_time.plot(kind = 'bar', x = 'commit_date_by_time', y = 'commit_sha', figsize = (10, 5))

fig.bar_label(fig.containers[0])
fig.plot()
```

# Most frequent words in commit message

```python
word_list = []
for i in df_cm['commit_message'].values.tolist():
  for word in list(map(str, str(i).split())):
    if word != 'nan':
      word_list.append(word.lower())
```

```python
df_msg = pd.DataFrame({'Word': word_list})
```

```python
df_msg['Word'].value_counts().head()
```

```python
file_list = []
for i in df_cm['commit_files_changed'].values.tolist():
  for file in list(map(str, i[1:-2:].split(','))):
    if len(file) > 0:
       file_list.append(file.lstrip())
```
