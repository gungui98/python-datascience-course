```python
user_names = []
date_times = []
files_change = []
#get the number of files has changed, the time committed, the name of user committed per commit
for pr in open_prs: 
    user_name = pr.user.login 
    repo_owner = user_name
    repo_name = 'python-datascience-course'
    commits = g.get_repo(f"{repo_owner}/{repo_name}").get_commits()
    for com in commits: 
        date_time = com.commit.author.date 
        file_change = len(com.files)
        user_names.append(user_name) 
        date_times.append(date_time)
        files_change.append(file_change)
```

```python
#check the data
for i in range(0, len(user_names)): 
    print(user_names[i], " ", date_times[i], " ", files_change[i])
```

```python
#build dataFrame
Data = []; 
for i in range (0, len(user_names)): 
    row = [user_names[i], date_times[i], files_change[i]]
    Data.append(row)
df = pd.DataFrame(data = Data, columns = ['UserName', 'DateTime', 'FilesChange'])
df.head()
```

```python
#put data to file.csv
df.to_csv('AI_K67_pythondata.csv', index = False)
```
