```python
import pandas as pd
```

```python
df = pd.read_csv("matplotlib_commits.csv")
df_files = pd.read_csv("matplotlib_file_id.csv")
```

```python
df = df.loc[pd.to_datetime(df["Commit time"]) >= pd.to_datetime("2022-01-01 00:00:00")].reset_index(drop=True)
df
```

```python
df = df.drop_duplicates()
df
```

```python
condition = True
for col in df.columns:
    condition = condition & df[col].notna()
df = df[condition]
df
```

```python
df.to_csv("matplotlib_commits_cleaned.csv", index=False)
```
