```python
from github import Github
import csv
```

```python
# Create github instance
g = Github("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")

# Specify the repository details
repo_owner = 'microsoft'
repo_name = 'sql-server-samples'

# Get the repository object
repo = g.get_repo(f"{repo_owner}/{repo_name}")

# Get commits from repository
commits = repo.get_commits()
```

```python
# Create list to store commit data
list_commit = []

# Extract commit information and add it to the commit_data list
for commit in commits:
    files_changed = len(commit.files)   # Number of files changed
    lines_changed = commit.stats.total  # Number of lines changed

    list_commit.append({
        'Commit ID': commit.sha,
        'Message': commit.commit.message,
        'Author Name': commit.commit.author.name,
        'Author Email': commit.commit.author.email,
        'Date': commit.commit.author.date,
        'Files Changed': files_changed,
        'Lines Changed': lines_changed
    })

# Write data in csv file
csv_file = 'data.csv'
fieldnames = list_commit[0].keys()

with open(csv_file, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=fieldnames)
    writer.writeheader()
    writer.writerows(list_commit)
```
